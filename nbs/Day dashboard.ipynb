```python
import foodcosts.core as core
import pandas as pd
```

```python
shop = 'LATEM'
date = '2023-01-26'
```

# Shop LATEM

```python
def get_all_sold_items_for_given_day_and_shop(day, shop):    
    return core.get_query_res_as_pd(
        f"""
        SELECT 
            vp.name,
            sum(vpol.quantity) as qty_sold,
            sum(vpol.quantity*vp.price) as total
        FROM v_pos_order_line as vpol
        left join v_pos_order as vpo on vpo.id = vpol.order_id
        left join v_pos_session as vs on vs.id = vpo.session_id
        left join v_pos as vpos on vpos.id = vs.pos_id
        left join v_shop as vshop on vpos.shop_id = vshop.id
        left join v_product as vp on vp.id = vpol.product_id
        where vshop.name = '{shop}'
        and created_at between '{day} 00:00' and '{day} 23:59'
        group by vp.name
        order by total desc;
        """
    )
```

```python
all_sold_items_for_given_day_and_shop_df = get_all_sold_items_for_given_day_and_shop(date,shop)
```

```python
all_sold_items_for_given_day_and_shop_df.index = all_sold_items_for_given_day_and_shop_df.name
```

```python
def get_all_sold_items_for_given_day(day):    
    return core.get_query_res_as_pd(
        f"""
        SELECT 
            vp.name,
            sum(vpol.quantity) as qty_sold,
            sum(vpol.quantity*vp.price) as total
        FROM v_pos_order_line as vpol
        left join v_pos_order as vpo on vpo.id = vpol.order_id
        left join v_pos_session as vs on vs.id = vpo.session_id
        left join v_pos as vpos on vpos.id = vs.pos_id
        left join v_shop as vshop on vpos.shop_id = vshop.id
        left join v_product as vp on vp.id = vpol.product_id
        and created_at between '{day} 00:00' and '{day} 23:59'
        group by vp.name
        order by total desc;
        """
    )
```

```python
all_sold_items_for_given_day_df = get_all_sold_items_for_given_day(date)
```

```python
all_sold_items_for_given_day_df.index = all_sold_items_for_given_day_df.name
all_sold_items_for_given_day_df['all_qty_sold'] = all_sold_items_for_given_day_df.qty_sold
all_sold_items_for_given_day_df['all_total'] = all_sold_items_for_given_day_df.total
all_sold_items_for_given_day_df.drop('qty_sold', axis=1, inplace=True)
all_sold_items_for_given_day_df.drop('total', axis=1, inplace=True)
```
