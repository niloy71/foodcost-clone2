```python
def get_email_subject(date, day):
    return f"""{date}'s - {day} -report"""


def get_daily_report(date, day, total_sales, shops_total_sales, re_restaurants_total_sales, events_total_sales):
    total_sales = int(total_sales);
    shops_total_sales = int(shops_total_sales) 
    re_restaurants_total_sales = int(re_restaurants_total_sales)
    events_total_sales = int(events_total_sales)
    
    return f"""
# {date}'s - {day} -report

  

Today sales were {total_sales} EUR across the business:

  

- {shops_total_sales} EUR sold in stores

- {re_restaurants_total_sales} EUR sold in corporate restaurants

- {events_total_sales} EUR sold in events

        """


def get_store_sales_title():
    return f"""
## Store sales

### Stores sales per shop


        """


def get_shop_sales(shop_name, total_sale, weekday, avg_sales, change_pct, change):
    total_sale = int(total_sale)
    avg_sales = int(avg_sales)
    change_pct = round(change_pct, 1)
    
    return f"""
- {shop_name} sold {total_sale} EUR. Average of last 4 {weekday}’s was {avg_sales} EUR. That’s {change_pct}% {change} vs. average of last 4 {weekday}'s

        """


def get_product_stock_out_title(outlet_type):
    return f"""
### Products that were likely in stock out

  

The products that were sold in the each days of last 7 days for a given {outlet_type} but didn’t sell today:

 
        """


def get_stock_out_product_details(product_name, outlet):
    return f"""
- {product_name} in {outlet}

        """


def get_corporate_restaurant_sales_title():
    return f"""
## Corporate restaurant sales

  

### Sales per corporate restaurant

  
  """


def get_corporate_restaurant_sales(name, total_sale, weekday, avg_sales, change_pct, change):
    total_sale = int(total_sale)
    avg_sales = int(avg_sales)
    change_pct = round(change_pct, 1)
    
    return f"""
- {name} sold {total_sale} EUR. Average of last 4 {weekday}’s was {avg_sales} EUR. That’s {change_pct}% {change} vs. average of last 4 {weekday}'s

        """
def get_no_sales_in_corporate_restaurant_md():

    return f"""
There is no sales in the corporate restaurant today

        """
```

```python
def get_all_shop_sales(start_date, end_date):
    query = f"""
        SELECT 
            v_shop.id as id,
            v_shop.name as name,
            day,
            total
        FROM v_shop_sales_statistics
        LEFT JOIN 
            v_shop on v_shop.id = v_shop_sales_statistics.shop_id 
        WHERE 
            v_shop.name IS NOT NULL AND   
        day BETWEEN '{start_date}' and '{end_date}';
        """

    return core.get_query_res_as_pd(query)


def get_all_corporate_restaurant_sales(start_date, end_date):
    query = f"""
        SELECT 
            v_resto_entreprise.id as id,
            v_resto_entreprise.name as name,
            day,
            total
        FROM v_shop_sales_statistics
        LEFT JOIN 
            v_resto_entreprise on v_resto_entreprise.id = v_shop_sales_statistics.shop_id 
        WHERE 
        v_resto_entreprise.name IS NOT NULL AND     
        day BETWEEN '{start_date}' and '{end_date}';
        """

    return core.get_query_res_as_pd(query)


def get_all_event_sales(start_date, end_date):
    query = f"""
        SELECT SUM(amount_total) as total
        FROM sale_order 
        WHERE 
        state = 'sale' AND date_order BETWEEN '{start_date}' and '{end_date}';
        """

    return core.get_query_res_as_pd(query)
```

```python
def get_sales_details_html(sales_details_md, change):
    sales_details_html = markdown.markdown(sales_details_md)
    if(change == "increase"):
        return sales_details_html[:9] + f"""<span style="background-color:#00FF00">""" + sales_details_html[9:-11] + f"""</span>""" + sales_details_html[-11:];
    elif(change == "decrease"):
        return sales_details_html[:9] + f"""<span style="background-color:#FFA07A">""" + sales_details_html[9:-11] + f"""</span>""" + sales_details_html[-11:];
    else:
        return sales_details_html;
    
```

```python
def compare_value_in_percentage(x, y):
    try:
        return round((x*100/(y)), 2)
    except ZeroDivisionError:
        return float('NaN')
```

```python
def get_change_percentage(x, y):
    change_percentage = compare_value_in_percentage(x, y)
    change = "decrease" if change_percentage < 0 else "increase"
    abs_change_percentage = abs(change_percentage)

    return change, abs_change_percentage
```

```python
# Creating Markdown for daily report

daily_report = get_daily_report(
    today,
    weekday,
    total_sales,
    total_sales_at_shops,
    total_sales_at_corporate_restaurants,
    total_sales_at_events)

# Store sales
daily_report = daily_report + get_store_sales_title()
daily_report_html = markdown.markdown(daily_report)

# Stores sales per shop
for shop in shop_list:
    sale_by_shop = all_shop_sales[all_shop_sales['name']
                                  == shop]['total'].values.sum()
    sale_by_shop_in_last_four_weeks_at_this_weekday = all_shop_sales_for_last_four_weeks_at_this_weekday[
        all_shop_sales_for_last_four_weeks_at_this_weekday['name'] == shop]['total'].values.sum()
    avg_sale_at_this_weekday_in_last_four_week = round(
        (sale_by_shop_in_last_four_weeks_at_this_weekday / 4), 2)

    difference_in_sale = sale_by_shop - avg_sale_at_this_weekday_in_last_four_week

    change, change_percentage = get_change_percentage(
        difference_in_sale, avg_sale_at_this_weekday_in_last_four_week)

    shop_sales_md = get_shop_sales(shop, sale_by_shop, weekday,
                       avg_sale_at_this_weekday_in_last_four_week, change_percentage, change)
    
    daily_report += shop_sales_md
    daily_report_html +=  get_sales_details_html(shop_sales_md, change)


# Products that were likely in stock out
daily_report += get_product_stock_out_title("shop")
daily_report_html +=  markdown.markdown(get_product_stock_out_title("shop"))

for shop in shop_list:
    products_sold_for_shops = product_sold_for_shops[shop]
    sales_frequency = products_sold_for_shops['product_name'].value_counts()

    for product, count in sales_frequency.items():
        if (count == 7):
            sales_today = products_sold_for_shops.loc[(products_sold_for_shops['product_name'] == product) & (
                pd.to_datetime(products_sold_for_shops['day']) == pd.to_datetime(today))]
            if (len(sales_today) == 0):
                stock_out_product_details = get_stock_out_product_details(product, shop)
                daily_report += stock_out_product_details
                daily_report_html +=  markdown.markdown(stock_out_product_details)


# Corporate restaurant sales
# Sales per corporate restaurant
daily_report += get_corporate_restaurant_sales_title()
daily_report_html +=  markdown.markdown(get_corporate_restaurant_sales_title())

if(total_sales_at_corporate_restaurants == 0):
    daily_report += get_no_sales_in_corporate_restaurant_md()
    daily_report_html +=  markdown.markdown(get_no_sales_in_corporate_restaurant_md())
else:
    for re_restaurant in re_restaurant_list:
        sale_by_corporate_restaurant = all_corporate_restaurant_sales[all_corporate_restaurant_sales['name']
                                                                      == re_restaurant]['total'].values.sum()
        sale_by_corporate_restaurant_in_last_four_weeks_at_this_weekday = all_corporate_restaurant_sales_for_last_four_weeks[
            all_corporate_restaurant_sales_for_last_four_weeks['name'] == re_restaurant]['total'].values.sum()
        avg_sale_at_this_weekday_in_last_four_week = round(
            (sale_by_corporate_restaurant_in_last_four_weeks_at_this_weekday / 4), 2)
        
        difference_in_sale = sale_by_corporate_restaurant - \
            avg_sale_at_this_weekday_in_last_four_week
        
        change, change_percentage = get_change_percentage(
            difference_in_sale, avg_sale_at_this_weekday_in_last_four_week)

        corporate_restaurant_sales = get_corporate_restaurant_sales(re_restaurant, sale_by_corporate_restaurant, weekday,
                                           avg_sale_at_this_weekday_in_last_four_week, change_percentage, change)
        
        daily_report += corporate_restaurant_sales
        daily_report_html +=  get_sales_details_html(corporate_restaurant_sales, change)

    # Products that were likely in stock out
    daily_report += get_product_stock_out_title("corporate restaurant")
    daily_report_html +=  markdown.markdown(get_product_stock_out_title("corporate restaurant"))
    
    for restaurant_name in re_restaurant_list:
        products_sold_for_corporate_restaurant = product_sold_for_corporate_restaurants[
            restaurant_name]
        sales_frequency = products_sold_for_corporate_restaurant['product_name'].value_counts(
        )
    
        for product, count in sales_frequency.items():
            if (count == 7):
                sales_today = products_sold_for_corporate_restaurant.loc[(products_sold_for_corporate_restaurant['product_name'] == product) & (
                    pd.to_datetime(products_sold_for_corporate_restaurant['day']) == pd.to_datetime(today))]
                if (len(sales_today) == 0):
                    stock_out_product_details = get_stock_out_product_details(product, restaurant_name)
                    
                    daily_report += stock_out_product_details
                    daily_report_html +=  markdown.markdown(stock_out_product_details)

                
# print(daily_report_html)
display_markdown(daily_report, raw=True)
```
